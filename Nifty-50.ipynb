# Downloading the data for analysis


```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

tickers = [
"^NSEI",
"^NSEBANK",
"^CNXAUTO" ,
"^CNXFMCG",
"^CNXIT",
"^CNXMEDIA",
"^CNXMETAL",
"^CNXPSE",
"^CNXREALTY",
"^CNXENERGY",
"^CNXINFRA",
"^CNXSERVICE",
"^CNXMNC",
]  # Add more sector indices as needed

data = {}
for ticker in tickers:
    data[ticker] = yf.download(ticker, start="2014-01-01", end="2024-01-01")
```

## The volume for the NIFTY-50 stocks are too uncertain, but there are increasing from 2014 to mid 2020. Its maxima is at between 2020-2021 , seems because of the low prices of stocks due to covid-19, people bought more stocks due to the fall of the market and lower prices.

```python
# Calculate rolling average of volume to smooth out the data
df_Nifty50['Volume 30 Day MA'] = df_Nifty50['Volume'].rolling(window=30).mean()

# Plot the volume with rolling average
plt.figure(figsize=(14, 7))
plt.plot(df_Nifty50['Volume'], label='Nifty 50 Volume', alpha=0.5)
plt.plot(df_Nifty50['Volume 30 Day MA'], label='30 Day Moving Average of Volume', color='red')
plt.title('Nifty 50 Trading Volume with 30 Day Moving Average (2014-2024)')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.grid(True)
plt.show()
```

# The above graph gives the following insights : -

# Volume Trends: By examining the volume data over time, There are some maximas between mid of 2019 to 2020 , also around the third or fourth month of 2020.  High trading volumes often correspond to significant market events or news. The peaks are due to covid waves during that time ( a historical event )

```python
train = df_Nifty50.iloc[: int( len(df_Nifty50)*0.95 ) ]
test = df_Nifty50.iloc[int( len(df_Nifty50)*0.95 ):]

#
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX( train['Open'] ,  order = (1 ,  0  ,  0 ),  seasonal_order =(2, 1, 0 , 12) )

result = model.fit()
result.summary()

```

```python
start = len(train)
end = len(train) + len(test) - 1

# Predictions for one-year against the test set
predictions = result.predict(start, end,  typ = 'levels').rename("Predictions")

# plot predictions and actual values
predictions.plot(legend = True)
test['Open'].plot(legend = True)
```
